```python
import MetaTrader5 as mt5
from datetime import datetime
import pandas as pd
```

```python
login, password = open('credentials').read().split()
```

```python
mt5.initialize(login=int(login), password=password, server='demo.mt5.xpi.com.br:443')
```

```python
ticks = mt5.copy_ticks_range('PETR4',
                             datetime(2024,7,10),
                             datetime(2024,7,20),
                             mt5.COPY_TICKS_TRADE
                            )

df_ticks = pd.DataFrame(ticks)

df_ticks
```

```python
df_ticks['time'] = pd.to_datetime(df_ticks['time'], unit='s')
df_ticks['time_msc'] = pd.to_datetime(df_ticks['time_msc'], unit='ms')

df_ticks
```
